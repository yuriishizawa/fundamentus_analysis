```python
from selenium import webdriver
import pandas as pd

DRIVER_PATH = '/usr/bin/chromedriver'

driver = webdriver.Chrome(executable_path=DRIVER_PATH)
driver.implicitly_wait(5) # seconds
driver.get('https://www.fundamentus.com.br/buscaavancada.php')
```

```python
columns = ["Papel",'EV/EBIT', 'P/L', "P/VP",'Div.Yield', "Mrg. Líq.", "Patrim. Líq"]

buttons = {
    "pl_min": 0,
    "pl_max": 20,
    "firma_ebit_min": 0,
    "margemebit_min": 0,
    "patrim_min": 1e8
}
```

```python
for button in buttons:
    field = driver.find_element(by='name', value=button)
    field.click()
    field.send_keys(buttons[button])
```

```python
search_button = driver.find_element(by='xpath', value="/html/body/div[1]/div[2]/form/input")
search_button.click()
```

```python
driver.find_element(by='xpath', value='/html/body/div[1]/div[2]/table')
```

```python
df = pd.read_html(driver.page_source.replace(",", "."))[0][columns]
df['Papel'] = df["Papel"].apply(lambda x: x[:4])
df = df.drop_duplicates('Papel')
df['Div.Yield'] = df["Div.Yield"].str.replace("%", "").astype("float").div(100)
df['Mrg. Líq.'] = df["Mrg. Líq."].str.replace("%", "").astype("float").div(100)
```
